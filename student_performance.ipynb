```
#independent variables
x=data[['VisITedResources','raisedhands']]
```

```
#dependent variable
y=data['Class']
y=y.astype('int')
y
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```
from sklearn.linear_model import LogisticRegression 
logmodel=LogisticRegression(max_iter=1000)
```

```
logmodel
```

```
y_train=y_train.astype('int')
y_test=y_test.astype('int')
x_train=x_train.astype('int')
x_test=x_test.astype('int')
```

```
logmodel.fit(x_train,y_train)
predictions=logmodel.predict(x_test)
```

```
#checking for the accuracy score without svm
from sklearn.metrics import jaccard_similarity_score
accuracy_score=jaccard_similarity_score(y_test,predictions)
print(accuracy_score*100)
```

```
from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(y_test,predictions)
print(matrix)
```

```
sns.heatmap(matrix,square=True, annot=True, fmt='d',cbar=False,cmap='Blues')
```

```
X=np.asarray(x_train)
Y=np.asarray(y_train)
X=X.astype('int')
```

```
from sklearn import svm
```

```
svm_linear=svm.SVC(kernel="linear", gamma="scale",)
svm_linear.fit(X,Y)
```

```
predictions=svm_linear.predict(x_test)
```

```
#checking svm score with svm
accuracy_score=jaccard_similarity_score(y_test,predictions)
print(accuracy_score*100)
```

```
from mlxtend.plotting import plot_decision_regions


plot_decision_regions(X, Y, clf=svm_linear)  

 
plt.show()                
```

```
svm_rbf=svm.SVC(kernel="rbf", gamma="scale")
svm_rbf.fit(X,Y)
```

```
prediction1=svm_linear.predict(x_test)
accuracy_score=jaccard_similarity_score(y_test,prediction1)
print(accuracy_score*100)
```

```
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,Y, clf=svm_rbf)   
plt.show()                
```

```
svm_poly = svm.SVC(C=1.0, kernel='poly', degree=3, gamma=2,max_iter = 1e5)
svm_poly.fit(x_train,y_train)
```

```
prediction2=svm_poly.predict(x_test)
accuracy_score=jaccard_similarity_score(y_test,prediction2)
print(accuracy_score*100)
```

```
plot_decision_regions(X, Y, clf=svm_poly)   
plt.show()                
```
